## 4. 請利用 "TP53.txt" mRNA 序列, 將其正股(sense strain)序列的第 0, 1 and 2 個不同起始的序列, 給列印出來?

```python
fn = open("TP53.txt")
seq = list()
for nt in list(fn.read()):
    if nt != '\n':
        seq += nt
for frame in [0, 1, 2]:
    print ("== Frame-", frame, "==")
    print ("".join(seq[frame:]))
```

## [Homework 11] 請將 TP53.txt 的 mRNAs 正反兩個股(main strand and complementary strand)，可能的 6 個 reading frames 給列印出?

![](http://www.davidmoore.org.uk/21st_Century_Guidebook_to_Fungi_PLATINUM/Guidebook_images/Ch18_Fig17_searching_for_ORFs.jpg)

```python
fn = open('TP53.txt')
seq = ""
for frame in [0, 1, 2]:
    for nt in fn.read():
        if nt != '\n':
            seq += nt
    print("== Main strain Frame {} ==".format(frame))
    print(seq[frame:])
fn.close()

fn = open('TP53.txt')
c_seq = ""
complementary_table = {'A':'T', 'T': 'A', 'G':'C', 'C':'G'}
for nt in fn.read():
    if nt != '\n':
        c_seq += complementary_table[nt]
for frame in [0, 1, 2]:
    print("== Complementary strain Frame {} ==".format(frame))
    print (c_seq[::-1])
fn.close()
```

## 6. 請參考 Week 15 的 Homework 9 的 Dictionary 的寫法，利用 standard genetic codon table 將 'TP53.txt' 的 6 個 ORFs 轉譯(translation) 成氨基酸的序列

```python
bases = "TCAG"
codons = [a + b + c for a in bases for b in bases for c in bases]
amino_acids = 'FFLLSSSSYY**CC*WLLLLPPPPHHQQRRRRIIIMTTTTNNKKSSRRVVVVAAAADDEEGGGG'
codon_table = dict(zip(codons, amino_acids))

fn = open('TP53.txt')
seq = ""
for frame in [0, 1, 2]:
    for nt in fn.read():
        if nt != '\n':
            seq += nt
    print("== Main strain Frame {} ==".format(frame))
    aa = ""
    for pos in range(frame, len(seq), 3):
        if len(seq[pos+frame:pos+3+frame]) == 3:
            aa += codon_table[seq[pos:pos+3]]
    print (aa)
fn.close()

fn = open('TP53.txt')
c_seq = ""
complementary_table = {'A':'T', 'T': 'A', 'G':'C', 'C':'G'}
for frame in [0, 1, 2]:
    for nt in fn.read():
        if nt != '\n':
            c_seq += complementary_table[nt]
    print("== Complementary strain Frame {} ==".format(frame))
    aa = ""
    for pos in range(frame, len(seq), 3):
        if len(c_seq[pos+frame:pos+3+frame]) == 3:
            aa += codon_table[c_seq[pos:pos+3]]
    print (aa)
fn.close()
```

## 8. 請利用 define function 的方法，來改寫 [Homework 11] 處理 (Main and complementary strains) 兩次的問題?

```python
def reading_frames(strain, seq):
    for frame in [0, 1, 2]:
        print("== {} strain Frame {} ==".format(strain, frame))
        print(seq[frame:])

fn = open('TP53.txt')
complementary_table = {'A':'T', 'T': 'A', 'G':'C', 'C':'G'}
seq = ""
c_seq = ""
for nt in fn.read():
    if nt != '\n':
        seq += nt
        c_seq += complementary_table[nt]
reading_frames('Main', seq)
reading_frames('Complementary', c_seq[::-1])
fn.close()
```

## [Homework 12] 請利用自型定義函數 (define function) 的方法，改寫將 'TP53.txt' 的 6 個 ORFs 轉譯 (translation) 成氨基酸序列的程式。

```python
def translation(strain, seq):
    bases = "TCAG"
    codons = [a + b + c for a in bases for b in bases for c in bases]
    amino_acids = 'FFLLSSSSYY**CC*WLLLLPPPPHHQQRRRRIIIMTTTTNNKKSSRRVVVVAAAADDEEGGGG'
    codon_table = dict(zip(codons, amino_acids))

    for frame in [0, 1, 2]:
        print("== {} strain Frame {} ==".format(strain, frame))
        aa = ""
        for pos in range(frame, len(seq), 3):
            if len(seq[pos+frame:pos+3+frame]) == 3:
                aa += codon_table[seq[pos:pos+3]]
        print(aa)
     
fn = open('TP53.txt')
complementary_table = {'A':'T', 'T': 'A', 'G':'C', 'C':'G'}
seq = ""
c_seq = ""
for nt in fn.read():
    if nt != '\n':
        seq += nt
        c_seq += complementary_table[nt]
translation('Main', seq)
translation('Complementary', c_seq[::-1])
fn.close()
```
