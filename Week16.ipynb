## 3. 請利用 "TP53.txt" mRNA 序列, 將其正股(sense strain)序列的第 0, 1 and 2 個不同起始的序列, 給列印出來?

```python
fn = open("TP53.txt")
seq = list()
for nt in list(fn.read()):
    if (nt =='A') or (nt =='T') or (nt =='G') or (nt =='C'):
        seq.append(nt)
for frame in [0, 1, 2]:
    print ("== Frame-", frame, "==")
    print ("".join(seq[frame:]))
```

## [Homework 10] 將 TP53.txt mRNA 序列，取得其反股(complementary strain) 的序列? (由 5'- > -3')

![](https://www.bioinformatics.org/tutorial/images/dnasequence.gif)

```python
fn = open('TP53.txt')
compl = {'A':'T', 'T': 'A', 'G':'C', 'C':'G'}
seq = list()
for nt in fn.read():
    if (nt == 'A') or (nt == 'T') or (nt == 'G') or (nt == 'C'):
        seq.append(compl[nt])
seq[::-1]
```
